```python
import pandas as pd

df = pd.read_csv("../data/cleaned/cleaned_movies.csv")

# Encode Certificate
cert_map = {'U':0, 'UA':1, 'A':2, 'S':3}
df['Certificate_Num'] = df['Certificate'].map(cert_map)

# Director Reputation (average gross per director)
director_avg = df.groupby('Director')['Gross'].mean()
df['Director_Reputation'] = df['Director'].map(director_avg)

# Star Power (sum of average gross per star)
stars = ['Star1','Star2','Star3','Star4']
for star in stars:
    star_avg = df.groupby(star)['Gross'].mean()
    df[star+'_Rep'] = df[star].map(star_avg)

# Combine star power into one metric
df['Star_Power'] = df[['Star1_Rep','Star2_Rep','Star3_Rep','Star4_Rep']].sum(axis=1)

# Keep numeric features
feature_cols = ['Runtime','IMDB_Rating','Meta_score','No_of_Votes',
                'Certificate_Num','Director_Reputation','Star_Power','Released_Year']

X = df[feature_cols]
y = df['Gross']

X.head()
```
